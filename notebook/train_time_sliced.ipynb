````python
df = pd.read_csv("/home/mdafifal.mamun/notebooks/triagerX/notebook/data/openj9/merged_data_dated.csv")
````

````python
import re

codes = []

for idx, row in df.iterrows():
    issue_body = row["issue_body"]
    
    if type(issue_body) == str:
        code_blocks = re.findall(r'```([\s\S]+?)```', issue_body)

        if len(code_blocks) > 0:
            codes.append(code_blocks)
````

````python
code = "\n".join(codes[0])
````

````python
toks = tokenizer(
    "",
    padding="max_length",
    max_length=512,
    truncation=True,
    return_tensors="pt",
)
````
